```python
## Class for Python/SQL functions
## Used for extracting data from characterStats excel database 
## structure, then to excel for functioning safely
class SQLExtraction(object):
  ''' how to access all information how to pull data from the excel
  file to collect data when it is messy
  '''

  ## Using pandas and sqllite3 and *sqlalchemy(needs to be
  ## to python scripting) on built in python
  import pandas as pd
  import sqllite3
  from sqlalchemy import create_engine

  def WriteNewTable():

    ## defining a few variables
    ## defining an input and output
    file = 'Statsofclass-attributes.xlsx'
    output = 'use.xlsx'

    ##Creating in memory, for safer access and easily getting rid of it
    engine = create_engine('sqllite://', echo=False)
    ##Creating dataframe with sheet name (IMPORTANT TO SAVE EXCEL ON
    ##CHARACTER STATS!(with sheetname=excel and 'chractersinfo'=sqldatabase name))
    df = pd.read_excel(file, sheet_name='Characters')
    df.to_sql('charactersinfo', engine, if_exists='replace', index=False)
  
    ##results of the table that can read the info quickly
    results = engine.execute('Select  * from chractersinfo')

    final = pd.DataFrame(results, columns = df.columns)
    final.to_excel(output, index=False)





```
