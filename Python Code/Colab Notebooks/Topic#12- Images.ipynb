```python
## We can use numpy to create our image
import numpy as np

#  We create an image of 512 by 512 containing only zeros.
w, h = 512, 512
data = np.full((h, w), 0, dtype=np.uint8)

## 0 is black and 255 is white
```

```python
# Now we can 2D array in a .png

from PIL import Image

img = Image.fromarray(data, 'L')
img.save('my_graysclae_img.png')
```

```python
## Now if we want images in color
## First 0 in data is red, second 0 in data is green, third 0 is blue
w, h = 512, 512
data = np.full((h, w, 3), np.array([0, 0, 0]), dtype=np.uint8)

## This is for getting the array, and save it
img = Image.fromarray(data, 'RGB')
img.save('my_color_img.png')

## We can mix the colors to create new colors too
```

```python
## We can take an existing image and modify it
## First downland an image (mont-saint-michel.png)
## We open it and make it into an array
import numpy as np
from PIL import Image
img = Image.open("mont-saint-michel.png")
data = np.array(img)
```

```python
## Creating a mirror image of the image, halfway up, (by dividing by 2)
## Then using two for loops for x and y, top pixel is equal to the same color
## down

def do_something_fancy(img):
       height = img.shape[0]
       width = img.shape[1]

       mirror_point = int(height/2)

       for x in range(width):
           for y in range(mirror_point):
            top_pixel = img[y, x]
            img[height-y-1,x] = top_pixel

do_something_fancy(data)
img2 = Image.fromarray(data, 'RGB')
img2.save('my_new_img.png')
```

```python
#It is possible to check all contours in the image
## We detect the luminance (brightness in the function)

## If there is a huge gap from 255 to 0, there is a contour

def luminance(pixel):
    r = pixel[0]
    g = pixel[1]
    b = pixel[2]
    return (r+g+b)/3

    
```

```python
## Now we implement the contour detection

def edgedetect(img):
    height = img.shape[0]
    width = img.shape[1]

    for x in range(width):
        for y in range(height):
            if y < height-1 and x < width-1:
              botrt = img[y+1, x+1]
              thislum = luminance(img[y,x])
              brlum = luminance(botrt)
              ## We can change the last digit, to modify the contours
              ## in the image (pixels size)
              ## and the 0, 0, 0 represents black, and 255, 255, 255 is black
            if abs(brlum-thislum) > 10:
                img[y,x] = [0,0,0]
            if abs(brlum-thislum) <= 10:
                img[y,x] = [255,255,255]
                

## Now we try it on our previous picture
img = Image.open("mont-saint-michel.png")
data = np.array(img)

edgedetect(data)
img2 = Image.fromarray(data, 'RGB')
img2.save('contour.png')
```
