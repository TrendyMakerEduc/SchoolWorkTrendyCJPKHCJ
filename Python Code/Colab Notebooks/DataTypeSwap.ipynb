```python
##Integer Input to check if input is digit or alphabet letter

def ConversionInput():
##First we assign the input variable
  inter = (input("Please enter a number or name: "))
  ##Using the built in isdigit(), we can check if the string contains a number
  if(inter.isdigit()):
    inter = int(inter)
    print(type(inter))
  ##Using the built in isalpha(), we can check if the string is a string
  elif(inter.isalpha()):
    print(type(inter))
##For certain situations, we must create the function ourselves, to check for value errors, to try to make situational advantages
##To functions, by trying to do the function, it will work as long as there is no ValueError!
  else:
    try:
      inter = float(inter)
      return type(inter)
    except ValueError:
      return type(inter)
##When would we use this? When making data classes, it can be quite useful to order the types, convert the types, so they can work
##With their own data using ordinal, nominal, ratios, and interval data, which is great to use to convert input types
##This can be especially useful when creating global input variables in a class!
ConversionInput()


```
