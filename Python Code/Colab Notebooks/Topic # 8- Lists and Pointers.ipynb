```python
## List using strings
L = ['Hi', 'How', 'Are', 'You', 'Doing', '?']
print(L[0:4])

## List using floats
F = [2.133, 2.345, 2.5546, 2.7589, 2.987]
print(F[0:4])

## List using int
I = [2,3,4,5,6,7,8]
print(I[0:4])

## Different types
Q = [2, 'at', 2.345, I]
print(Q[0:2])

## Variables in lists
Big = [L, F, I, Q]
print(Big[0:2])

## Every type can go in a list

## finding the number of elements for a list
print(len(Big))

## function for printing lists
def print_list(B):
  for item in B:
    print(item)

## Function for printing even element indexes
def print_even(l):
  position = 0
  for item in l:
    if position%2 == 0:
      print(item)
    position = position + 1

print_even(Big)

## Testing if 5 is inside the I list(Should return True)
5 in I

## Adding lists together
print(I + Q)

## Automatically making a list from 1 - 5, 5 excluded
list(range(1,5))

## Every number from 0 - 17
list(range(0,17))

## Every number from -10 - 0
list(range(-10, 0))

## Every number from 10 to 0 (counting down)
list(range(10, -1, -1))

##All even numbers from 1 - 20
list(range(0,21,2))
print_even(list(range(0,21)))

## Changing element 5 on a list
lister = list(range(0,25,1))
lister[4] = "x"
lister[0:2] = 10, 11
##removing elements
lister[3:6] = []
```

```python
## A changes as well, we are not creating a new variable b, it is creating another name for a but it is the same variable.. only with lists
## and objects like that

## This is creating an alias.. This saves on memory, because memory runs out very fast.

a = [1, 2, 3]
b = a
print(a, b)
b[0] = 10
print(a, b)

## If we want a real copy we different variables, we do this
c = a[:]
print(a, b, c)
c[0] = 32
print(a, b, c)

```
