```python
twoD = [['a', 'b', 'c'],
        ['d', 'e', 'f'],
        ['g', 'h', 'i']]
## First problem- Printing out contents in given row
def printRow(mat, row):
 for x in range(len(mat)):
    print(mat[row][x])
    x =+ 1
printRow(twoD, 1)

#Second Problem- Printing out contents in given column
def printCol(mat, col):
  for x in range(len(mat)):
    print(mat[x][col])
    x += 1
printCol(twoD, 2)

#Third Problem- Printing from top left to bottom right
def printDownRight(mat):
  RowIndex = 0
  for Column in range(len(mat)):
    print(mat[RowIndex][Column])
    RowIndex += 1
printDownRight(twoD)

#Fourth Problem- Printing upright from bottomleft to topright
def printUpRight(mat):
  ColumnIndex = 0
  Row = -1
  for ColumnIndex in range(len(mat)):
    print(mat[Row][ColumnIndex])
    Row -= 1

printUpRight(twoD)

# Fifth Problem- Printing from top right to bottom left
def printDownLeft(mat):
  ColumnIndex = -1
  Row = 0
  for Row in range(len(mat)):
    print(mat[Row][ColumnIndex])
    ColumnIndex -= 1

printDownLeft(twoD)

# Sixth Problem- Printing from bottom right to Top left
def printUpLeft(mat):
  ColumnIndex = -1
  Row = -1
  for x in range(len(mat)):
    print(mat[Row][ColumnIndex])
    Row -= 1
    ColumnIndex -= 1
    

printUpRight(twoD)

#testing on another 2D array

uhoh = [['a', 'b', 'c', 'w', 'e', 'q', 'e'],
        ['d', 'e', 'f', 'x', 'e', 'z', 'q'],
        ['g', 'h', 'i', 'y','e', 'w', 'w'],
        ['j', 'k', 'l', 'z','e', '3','3'],
        ['a', 'a', 'q', 'e','e','2','3']]

printUpLeft(uhoh)

## Its working!!
```
