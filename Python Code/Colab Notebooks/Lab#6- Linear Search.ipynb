```python
#3

def count_letters(char, fruit):
  count = 0
  for characters in fruit:
    if characters == char:
      count += 1
  return count

count_letters("a", "bannana" )
assert 3 == count_letters("a", "bannana")
#9

def remove_letter(char, string):
  replace = string.replace(char, "")
  return replace

assert "pple" == remove_letter("a", "apple")
assert "bnn" == remove_letter("a", "banana")
assert "banana" == remove_letter("z", "banana")
assert "Msssspp" == remove_letter("i", "Mississippi")
assert "" == remove_letter("b", "") 
assert "c" == remove_letter("b", "c")

# 10
def is_palindrome(word):
  return word == word[::-1]

assert(is_palindrome("abba"))
assert(not is_palindrome("abab"))
assert(is_palindrome("tenet"))
assert(not is_palindrome("banana"))
assert(is_palindrome("straw warts"))
assert(is_palindrome("a"))
# test(is_palindrome(""))    # Is an empty string a palindrome?

#11

def count(substring, string):
  count = 0
  increment = 0
  length = len(substring)
  for char in string[count: count + length]:
    if(substring in string):
      increment += 1
      print(increment)
    
    
  return increment

assert 2 == count("is", "Mississippi")
assert 2 == count("an", "banana")
#assert 2 == count("ana", "banana") 
#assert 1 == count("nana", "banana")
assert 0 == count("nanan", "banana")
#assert 4 == count("aaa", "aaaaaa")

#12

def remove(substring, string):
  new_string = string.replace(substring, "", 1)
  return new_string

assert "bana" == remove("an", "banana")
assert "bile" == remove("cyc", "bicycle")
assert "Missippi" == remove("iss", "Mississippi")
assert "bicycle" == remove("eggs", "bicycle") 

def remove_all(substring, string):
  new_string = string.replace(substring, "")
  return new_string

assert "ba" == remove_all("an", "banana")
assert "bile" == remove_all("cyc", "bicycle")
assert "Mippi" == remove_all("iss", "Mississippi")
assert "bicycle" == remove_all("eggs", "bicycle") 

def character_is_in_while(needle: str, haystack: str) -> bool:
  count = 0
  while (count <= len(haystack)):
    count += 1
    if(needle in haystack[count]):
      return True
    else:
      return False

assert True == character_is_in_while("a", "ban")

def character_is_in_for(needle: str, haystack: str) -> bool:
  for char in haystack:
    if(needle in haystack[:]):
      return True
    else:
      return False
    

assert True == character_is_in_for("a", "bfgdsgsdfan")


def character_is_at_while(needle: str, haystack: str) -> int:
  count = 0
  while (count <= len(haystack)):
    if(needle == haystack[count]):
      return count
    count += 1
assert 3 == character_is_at_while("a", "bngatewrt")   
    
def character_is_at_for(needle: str, haystack: str) -> int:
  count = 0
  for character in haystack:
    if needle == character:
      return count
    count += 1

assert 3 == character_is_at_for("a", "bngatewrt")

##KATTIS PROBLEMS


```
