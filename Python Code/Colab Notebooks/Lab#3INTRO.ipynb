```python
def Judging_Moose():
  data = input()       # Read a WHOLE, SINGLE line of input
  data = data.split()  # Split string into individual pieces
  l = int(data[0]) # Take string from data[X], convert it to int...
  r = int(data[1]) # ... And store it in some variable 
  total = l + r
  if(l > r):
    total = l * 2
    return "Odd " + str(total)
  elif (r > l):
    total = r * 2
    return "Odd " + str(total)
  elif(r == 0 and l == 0):
    return "Not a moose"
  else:
    return "Even " + str(total)

print(Judging_Moose())
    
    
```
