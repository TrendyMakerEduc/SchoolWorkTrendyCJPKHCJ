```python
import math
import csv
import numpy
from termcolor import colored


def load_asn2_data():
  """
  This function loads the file `wordle.csv` and returns a LIST of 5 letters words.
  We'll talk about LISTs formally in class in a few lectures, but maybe
  you can start guessing how they work based on what you see here...
  """
    
    
  reader = csv.reader(open('./wordle.csv', 'r'))        # wordle.csv MUST be in the same directory as this script for this to work!
  words = []  
  
  for r in reader:
    words.append(r[0]+r[1]+r[2]+r[3]+r[4])
        
  return words
```

```python
def color(char, color):
  """
  This function return a char in color.
  """
  if color == 'ORANGE':
    return '\033[93m' + char
  elif color == 'GREEN':
    return '\033[92m' + char
  else:
    return '\033[0m'+char
```

```python
def choose_a_word(list_of_words):
  """
  This function choose a word randomly from a list of word.
  """
  # INSERT YOUR CODE HERE
  ## Grabs random element based on load_asn2_data() function used in main function "wordle"
  List = numpy.random.choice(list_of_words , 1, replace=False)

  ## Converts the one list variable to a string
  Word = "".join([str(Words) for Words in List])
  
  ## Return the one word string
  return Word
  

```

```python
def give_hint(player_guess, word_to_guess):
  """
  This function gives hint to the user.
  Print the hints directly.
  """
  hints = ""
  index_position = 0
  # INSERT YOUR CODE HERE
  ## For initiating a loop in a range of 5 times
  for letter in range(len(player_guess)):

  ## Wordle Will be Restarted with a new word if letter is greater then 5, 
  ## and index_position is greater then 5 so that errors are not arising
    if index_position >= 5 or letter >= 5:
      print("Please input 5 letter words only.")
      index_position = 0
      return 
    

    ## Checks using the index_position variable, for letter guess and actual word and checks if the positions are the same
    ## and increments the variable after the three loops are done
    #### CHANGES COLOR OF TEXT TO GREEN
    elif player_guess[letter] in word_to_guess[index_position] and player_guess[letter] in player_guess[index_position]:
       hints += color(player_guess[letter], 'GREEN')
       
    ## Checks if letter is anywhere else in the word
    ## CHANGES COLOR OF TEXT TO ORANGE
    elif player_guess[letter] in word_to_guess :
     hints += color(player_guess[letter], 'ORANGE')

    ## If letter is not found in word or position, it returns a black color
    ## LEAVES TEXT AS DEFAULT (BLACK)
    else:
      hints += color(player_guess[letter], 'BLACK')

    ## Before loop goes to next letter, we increment the index position
    index_position += 1
    
    
  
  
  print(hints)
  print('\033[0m')

```

```python
def guess_word(word_to_guess):
  """
  This function as the player a word until they found the word or they guessed 5 times.
  Return True if the word is found, or False otherwise.
  """
  nb_guess = 0
  # INSERT YOUR CODE HERE

  ## Variables- Check_Conditions is always TRUE unless you win the game
  ## - GuessList and IncrementList are used together to increment a list and read till the last guess
  Check_Conditions = False
  GuessList = ["1st Guess", "2nd Guess", "3rd Guess", "4th Guess", "Last Guess"]
  IncrementList = 0

  ## Use while loop to check and loop for 5 guesses and no current win
  while nb_guess < 5 and Check_Conditions == False:

    ## Prints Guess, Guess takes input for letters, returns a hint from the text, and increments the guess count
    ## and List index
    print(GuessList[IncrementList])
    Guess = input("What is your  guess? (Please choose 5 latter words!) : ")
    give_hint(Guess, word_to_guess)
    nb_guess += 1
    IncrementList += 1

    ## If Guess is Corrent, You Win, and sets condition to True (through the wordle function)
    ## to Loop through to play the game again
    if Guess == word_to_guess :
      print("CONGRATULATIONS, YOU HAVE WON THE GAME!!!")
      Check_Conditions = True
      return Check_Conditions

    ## If not equal to the word to guess, keeps Check_Conditions as False to Loop through while loop again
    if Guess != word_to_guess:
      Check_Conditions = False

  
      ## if loop  for guesses are more then 5, returned value goes to wordle function, and prints Lost, and asks to Play Again
    if nb_guess >= 5:
      Check_Conditions = False
      return Check_Conditions
     
 


      
      

  
      
```

```python
def wordle():
  """
  This is the main function.
  This function load the data file, choose a word to guess and launch the game.
  """

  # INSERT YOUR CODE HERE
  # Set Load equal to choosing a word, and then Loading the one word string
  Load = choose_a_word(load_asn2_data())

  # Load the main game loop
  Check_Conditions = guess_word(Load)

  ## If the game has exceeded 5 guesses, or Lost
  if Check_Conditions == False:
    print("YOU HAVE LOST!!!")
    Play_Again = input("Play again? (y/n) ")

    ## If loop to take input as "y" to restart the game, through the wordle convienence function
    if Play_Again == "y" :
      wordle()

    ## All other character end the game loop
    else:
      print("Please come try again sometime!")
      return

  ## If the word has been correctly guessed, grabs returned true value, and prints the extra win message
  if Check_Conditions == True:
    print("YOU HAVE WON WOOHOO!!!")
    Play_Again = input("Play again? (y/n) ")

    ## If loop to take input as "y" to restart the game, through the wordle convienence function
    if Play_Again == "y" :
      wordle()

    ## All other character end the game loop
    else:
      print("Please come try again sometime!")
      return


    
  
```
