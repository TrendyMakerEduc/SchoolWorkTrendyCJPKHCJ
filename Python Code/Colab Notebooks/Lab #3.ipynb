```python
# Function to either Pass or Fail 0-100
## This Function must be used with type int


def grading_scheme():
  ## Converting input to type int
  grade = int(input("What is your grade?: "))
  
  ## Between 0 and 50
  if(grade < 50 and grade > 0):
    return print("FAILED")
  ## Between and equal to 50-100  
  if(grade >= 50 and grade <= 100):
    return print("PASS!")
  ## Less then 0  
  if(grade < 0):
    return print("Invalid grade number.")
  ## Greater then 100  
  if(grade > 100):  
    return print("Invalid grade number.")
  ## Equal to 0  
  if(grade == 0):
    return print("FAILED")
  ## Invalid Syntax for the odd reason 
  if(isinstance(grade, str)):  
    return print("Please put a valid number.") 
  else:
    return print("Please put in a valid number between 0-100.") 
  
grading_scheme()
```

```python
##Kattis Problem 1- Quadarant Selection

def Quadarant_Selection():
  x = int(input("What is the x value?: "))
  y = int(input("What is the y value?: "))
  ## Creating grid constraints
  while(x < -1000 or x > 1000):
    a = int(input("What is the x value?: "))
    x = a
  while(y < -1000 or y > 1000):
    a = int(input("What is the y value?: "))
    y = a
  ## Checking through all the quadarents
  if(x > 0 and y > 0):
    return print("1")
  if(x < 0 and y > 0):
    return print("2")
  if(x < 0 and y < 0):
    return print("3")
  if(x > 0 and y < 0):
    return print("4")


Quadarant_Selection()
```

```python
## Moose Antler Counting- Kattis Problem 2
## Counts for even tines on each side of the moose, and also odd pointed moose gets counted for as well as even

def tine_Count():
  l = int(input("What is the amount of points on the left side? : "))
  r = int(input("What is the amounts of points on the right side?: "))
  ## Creating tine constraints
  while(l >= 20 or l <= 0):
    a = int(input("What is the number on the left side of the moose? "))
    l = a
  while(r >= 20 or r <= 0):
    a = int(input("What is the number on the right side of the moose?: "))
    r = a
  
  if((l + r) % 2) == 0:
    print("Even" + " " + str(l + r))
  else:
    print("Odd" + " " + str(l + r))

tine_Count()
```
