```python
def scale_sim(s, min_s, max_s):
  value = (s - min_s) / (max_s - min_s)
  return abs(value)

scale_sim(12, 10, 20)
```

```python
def SMC(x,y):
  num11 = 0
  num10 = 0
  num00 = 0
  num01 = 0
  count = 0
  while count < len(x):

    if(x[count] == 1 and y[count] == 1):
      num11 += 1
    if(x[count] == 1 and y[count] == 0):
      num10 += 1
    if(x[count] == 0 and y[count] == 0):
      num00 += 1
    if(x[count] == 0 and y[count] == 1):
      num01 += 1
    if count == (len(x) - 1):
      SMC = ((num11 + num00) / (num11 + num10 + num00 + num01))
      return SMC
    if len(x) != len(y):
      return False
    count += 1
```

```python
def Jaccard(x, y):
  num11 = 0
  num10 = 0
  num01 = 0
  count = 0
  while count < len(x):

    if(x[count] == 1 and y[count] == 1):
      num11 += 1
    if(x[count] == 1 and y[count] == 0):
      num10 += 1
    if(x[count] == 0 and y[count] == 1):
      num01 += 1
    if count == (len(x) - 1):
      Jacc = (num11) / (num11 + num10 + num01)
      return Jacc
    if len(x) != len(y):
      return False
    count += 1
```

```python
import math
def Cosine(x, y):
  count = 0
  d1d2 = 0
  d1 = 0
  d2 = 0
  while count < len(x) or count < len(y):
    d1 += (x[count])**2
    d2 += (y[count])**2
    d1d2 += (x[count] * y[count])
    if count == (len(x) - 1):
      Formula = (d1d2 / ((d1**0.5) * (d2**0.5)))
      return Formula
    if len(x) != len(y):
      return False
    count += 1
```
