```python
##Set example
comp_sci = set({"Frank", "Reggie", "Bart", "Billy", "Reggie"})
print(comp_sci)
##The duplicates will get removed from this, and will replace the order based on the number of times each person showed up
#Will return True
print("Frank" in comp_sci)

#.union with the set will take the two sets and put them together
french = set({"Mercel", "Jessica", "Amanda", "Frank"})
new_set = comp_sci.union(french)
print(new_set)

#.intersection is where the two are common with each other
Common = comp_sci.intersection(french)
print(Common)

#.difference lets us see what is in what set, but not in an other
Diff = comp_sci.difference(french)
print(Diff)

```
