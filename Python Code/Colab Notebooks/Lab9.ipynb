```python
class Person:

  def __init__(self, first_name = "John", last_name = "Doe", email = "trendy@gmail.com"):
    self.first_name = first_name
    self.last_name = last_name
    self.email = email

  def get_first_name(self):
    return self.first_name

  def get_last_name(self):
    return self.last_name

  def get_email(self):
    return self.email

  def set_first_name(self, newFirstName):
    self.first_name = newFirstName
    return self.first_name

  def set_last_name(self, newLastName):
    self.first_name = newLastName
    return self.last_name

  def set_email(self, newEmail):
    self.first_name = newEmail
    return self.email
  
  def __repr__(self):
    return self.last_name + ", " + self.first_name + ":  " + self.email

  def __eq__(self, AnotherThing):
    return self.first_name == AnotherThing.first_name

```
