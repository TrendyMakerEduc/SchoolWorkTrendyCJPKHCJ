```python
## Question 1

def count_to_n_w(n):
  w = 0
  n = n - 1
  while w < n:
    w += 1
    print(w)
  
    
count_to_n_w(8)
```

```python
## Adds one to a a total of n times (Question 2)

def add_one_n_times_n(a, n):
  while a <= n or n != 0:
    a += (n / n)
    print(a)
    n = n - 1
  return a
add_one_n_times_n(7, 10)
```

```python
## Counting from 0 to n-1 using for and while loops

def count_to_n_f(n):
  for i in range(n):
      print(i)
      
    
count_to_n_f(8)
```
