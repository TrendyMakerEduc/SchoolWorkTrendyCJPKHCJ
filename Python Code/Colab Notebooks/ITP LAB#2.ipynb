```python
#7 Write a function called this_is_tough(n1, n2, n3, n4)

def this_is_tough(n1, n2, n3, n4):
  result1 = add_two_numbers_return(n1, n2)
  result2 = add_two_numbers_return(result1, n3)
  result3 = add_two_numbers_return(result2, n4)
  return result3

this_is_tough(1,2,3,4)
this_is_tough(3,6,9,12)
```
