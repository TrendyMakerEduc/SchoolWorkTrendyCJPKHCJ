```python
## Adding plus 1 to a variable based on the number you put in the x (lower) and y variable times, and return the value
def Adding(x, y):
  a = 1
  # Important that this is declared before the while loop
  y = y + x
  while(a < y):
   a += 1
   
  return a
  print(a)
Adding(5, 10)
```

```python
## Take n and multiply by answer, and then n goes down by 1

def do_stuff(n):
  answer = 1
  while n > 1:
    answer = answer * n
    n = n - 1
    print(answer, n)
  return answer
do_stuff(5)
```

```python
### Adding the sums of n from a number
def int_sum(n):
  x = 1
  y = n + x
  while n > 1 and n < y:
    x = x + n
    n = n - 1
    print(x, n)
  return x
    
int_sum(10)
    
```

```python
### Newtons Method
import math

## Set value of x between 0 and 1
def Formula(x):
  x = ((math.cos(x) - x ** 3.00) / (-math.sin(x) - 3 * x ** 2.00))
  return x
Formula(0.5)

def calculate_solution(x):
  y = Formula(x)
  while x != y :
    x = y
    y = Formula(x)
  return x
  
  

calculate_solution(0.5)


### REDO THIS ONE, should be between 0 and 1!
```
