```python
## CAR RENTAL APPLICATION


def calculate_total_charge(days, age, C, starting_odo, ending_odo):
  dollars = 0
  ageCost = 0
  Kilometers = ending_odo - starting_odo
  Average_Kms = (Kilometers - 100 * days) / days 
  
  ##Calculation for classification "B" 
  if(C == "B"):
    dollars = (days * 20) + (Kilometers * 0.3)
    if(age < 25):
      ageCost = days * 10
  ##Calculation for classification "D"
  elif(C == "D" and Average_Kms > 100):
    dollars = (days * 50) + (Average_Kms * 0.3)
    if(age < 25):
      ageCost = days * 10
  ##Secondary Calculation for classification "D"
  elif(C == "D"):
    dollars = (days * 50)
    if(age < 25):
      ageCost = days * 10
  else:
    return "That is an invalid Classification"
  dollars = dollars + ageCost
  return dollars
print(calculate_total_charge(2, 30, "D", 0, 500))
assert 20 == calculate_total_charge(1, 30, "B", 0, 0)
assert 50 == calculate_total_charge(1, 30, "D", 0, 0)
assert 30 == calculate_total_charge(1, 20, "B", 0, 0)
assert 60 == calculate_total_charge(1, 20, "D", 0, 0)
assert 50 == calculate_total_charge(1, 30, "B", 0, 100)
assert 50 == calculate_total_charge(1, 30, "D", 0, 100)
assert 60 == calculate_total_charge(1, 20, "B", 0, 100)
assert 60 == calculate_total_charge(1, 20, "D", 0, 100)
assert 190 == calculate_total_charge(2, 30, "B", 0, 500)
assert 145 == calculate_total_charge(2, 30, "D", 0, 500)
assert 210 == calculate_total_charge(2, 20, "B", 0, 500)
assert 165 == calculate_total_charge(2, 20, "D", 0, 500)
```
