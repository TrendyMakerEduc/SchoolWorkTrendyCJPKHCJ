```python
## We can use numpy to create our image
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from random import randint
from os import system, name
  
# import sleep to show output for some time period
from time import sleep
  
# define our clear function
def clear():
  
    # for windows
    if name == 'nt':
        _ = system('cls')
  
    # for mac and linux(here, os.name is 'posix')
    else:
        _ = system('clear')
  
  


characterX = 1 # Horizontal- this is the row number
characterY = 1 # Vertical- This is the column row
character = "A" # Character Class Grabbed from Save File
characterHealth = 100 #Character Health *important for stats
CharacterHealthLost = 0 #Health Lost *important for stats
Monsters ="Bai zhu0.png" # Monster List for changing variable
characterPosition = "| {0}  | ".format(character) #Labels the character selected
CharacterItems = [] # Character Items - stored in and list and appended
EnemyHitPoints = 0 #Enemy HitPoints on Board
EnemyHitPointsLost = 0 #EnemyHitPointsLost



GameBoard = [["|    | " for L in range(8)] for W in range(5)] #Defining the grid range

GameBoard[characterX][characterY] = characterPosition #Character positions

def MovingAround():  #Function for grid labels for monster, item, random events, chance for spells, etc
  while(characterX == 0 and characterY == 1):
    global characterHealth
    Monsters = "slimeingrassright08.png"
    img = cv2.imread(Monsters)
    width = 500
    height = 200 # keep original height
    dim = (width, height)
    EnemyHitPints = 20
    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    cv2_imshow(resized)
    print("                                You see a A grass bug up ahead, what would you like to do?")
    print("                                H = Hit, M = Magic, D = Defend, R = Run away")
    characterDecision = input()
     
    if(characterDecision == "H"):

      

      
      print("                                           You have missed")
      CharacterHealthLost = randint(0, 3)
      print("                                           You have lost {0} hitpoints from the buzzing attacks!".format(CharacterHealthLost))
      characterHealth -= CharacterHealthLost
      print("                                           CURRENT HEALTH: " + str(characterHealth))

    elif(characterDecision == "R"):
      print("                        You have successfully run away!")
      clear()
      break

  while(characterX == 0 and characterY == 0):
    global characterItems
    Monsters = "slime20.png"
    img = cv2.imread(Monsters)
    width = 500
    height = 200 # keep original height
    dim = (width, height)

    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    cv2_imshow(resized)
    print("                                You see a A PURPLE VIAL, what would you like to do?")
    print("                                PICK UP - P    WALK BY:   W")
    characterDecision = input()
     
    if(characterDecision == "P"):
      
      print("                                           You have picked up the PURPLE VIAL")
      CharacterItems.append("Purple Vial")
      
      break

    elif(characterDecision == "W"):
      print("                        YOU WALKED PAST THE VIAL")
      break

while True:
  for P in GameBoard:
    print("......  ......  ......  ......  ......  ......  ......  ......")
    print(" ".join(P))
    print("......  ......  ......  ......  ......  ......  ......  ......")
  print("Directions To the Dungeon")
  print("Up: W")
  print("Down: S")
  print("Left: A")
  print("Right: D")
  MovingAround()

  CharacterDirection = input("Please enter which way you would like to proceed...")
  if CharacterDirection == "W": #Moving Up
    GameBoard[characterX][characterY] = "|    | "
    characterX -= 1
    GameBoard[characterX][characterY] = "|  {0} | ".format(character)
    

  elif CharacterDirection == "S": #Moving DOWN
    GameBoard[characterX][characterY] = "|    | "
    characterX += 1
    GameBoard[characterX][characterY] = "|  {0} | ".format(character)
    

  elif CharacterDirection == "A": #Moving LEFT
    GameBoard[characterX][characterY] = "|    | "
    characterY -= 1
    GameBoard[characterX][characterY] = "|  {0} | ".format(character)
   

  elif CharacterDirection == "D": #Moving RIGHT
    GameBoard[characterX][characterY] = "|    | "
    characterY += 1
    GameBoard[characterX][characterY] = "|  {0} | ".format(character)
    
    
    
     
                                                        
```
